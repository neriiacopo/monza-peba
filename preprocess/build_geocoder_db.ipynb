```python
import pandas as pd
```

```python
pois = pd.read_csv("data/pois_osm.csv", sep=";", encoding="utf-8")
pois["type"] = "poi"

pois.head()
```

```python
addresses = pd.read_csv("data/COMUNE_MONZA_Numerazione_Civica_Comunale_4326.csv", sep=",", encoding="utf-8")

addresses["type"] ="adr"

addresses.head()
```

```python
joined = pd.concat([pois, addresses], ignore_index=True)
joined.drop(["id", "source"], axis=1, inplace=True)


joined.to_json("geocoder_db.json", orient="records", indent=2, force_ascii=False)
```
